2. The entire string is reversed. Example: `HELLO`$\rightarrow$`OLLEH`.
3. Replace any `" "` by `"*"`. Example: `I AM A SPY` $\rightarrow$ `I*AM*A*SPY`.
4. Characters at even positions (0-indexing) are lowercase. Example: `SECRET*TEXT`$\rightarrow$`sEcReT*TeXt`.
5. Add the spy-marker `"\#\#\#"` at the beginning and end of the message so we know where it starts and ends. Example: `SPY`$\rightarrow$`"\#\#\#SPY\#\#\#"`.

```python
msg_to_encode = msg1


#Check function from part A without print statements
def is_valid_message(msg):
    msg_to_encode = msg.replace(" ", "")
    if msg_to_encode.isupper() and msg_to_encode.isalpha():
        return True
    else:
        return False



def encode_message(msg):
    if not is_valid_message(msg):
        raise ValueError("Message must contain only uppercase letters and spaces.")
    else:
        shifted_msg = ""
        for char in msg:
            shifted_msg += chr((ord(char) - 65 + 3) % 26 + 65) if char != " " else " "

        reversed_shifted_msg = shifted_msg[::-1]

        asterix_msg = reversed_shifted_msg.replace(" ", "*")

        styled_msg = "".join(
            ch.lower() if i % 2 == 0 else ch.upper()
            for i, ch in enumerate(asterix_msg)
        )

        encoded_msg = "\"\#\#\#" + styled_msg + "\#\#\#\""
        
        return encoded_msg


print(f"Encoded message: {encode_message(msg_to_encode)}")
```

# Part D (2 marks)

A fellow spy has told you that it is possible to use Python's `map` method or list comprehensions to decode messages using the mirror dictionary `mirror_dict`. 

Write two alternative functions `decode_mirror_map` and `decode_mirror_listcomp` that use `map` to decode messages using the (possibly corrected) mirror dictionary `mirror_dict`. Check that these functions give the same results as in Part C when applied to `msg2`.

```python
def decode_char(ch, mirror_dict):
    if ch != " ":
        return mirror_dict.get(ch)
    else:
        return " "

def decode_mirror_map(message, mirror_dict):
    return "".join(map(lambda ch: decode_char(ch, mirror_dict), message))

print(f"Decoded map method: {decode_mirror_map(msg2, mirror_dict)}")

def decode_mirror_list_comp(message, mirror_dict):
    return "".join([decode_char(ch, mirror_dict) for ch in message])

print(f"Decoded list comprehension method: {decode_mirror_list_comp(msg2, mirror_dict)}")
```

# Part E (3 marks)

The message string `msg3` contains words separated by `"*"` that have been encoded using an unknown method. You are also given a list of important encoded keywords `important_words`. 

# E(i) (1 mark)

Create a dictionary `word_freq` with keys being the encoded words in the message and values the frequency of each word in the message. Print the word frequencies preceded by an informative print statement.

```python


def word_freq(message, important_words):
    words = message.replace("*", " ").split()
    freq_dict = {word: 0 for word in important_words}
    for word in words:
        if word in freq_dict:
            freq_dict[word] += 1

    return freq_dict

print(f"Word frequencies: {word_freq(msg3, important_words)}")
```

## E(ii) (1 mark)

Identify the important keywords in the message. Print these important keywords in alphabetical order, preceded by an informative print statement.

```python
def get_key_words(message, important_words):
    words = message.replace("*", " ").split()
    found_words = {word for word in words if word in important_words}
    return sorted(found_words)

print(f"Important words found in sorted order: {get_key_words(msg3, important_words)}")
```
